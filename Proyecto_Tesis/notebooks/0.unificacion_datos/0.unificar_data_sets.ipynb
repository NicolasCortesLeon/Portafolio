```python
import pandas as pd
```

```python
df1 = pd.read_csv("/Users/dayth/OneDrive/Escritorio/Proyectos/EDA + ML/data/rawdata/pag1.csv", encoding="utf-8", delimiter=",")
idx = df1.columns.get_loc('superficie_util')
df1.insert(idx, 'superficie_construida', df1['superficie_total'] - df1['superficie_util'])
df1 = df1.drop(columns=["url", "id", "descripcion", 'superficie_util'])
df1['estacionamientos'] = df1['estacionamientos'].apply(lambda x: 1 if pd.to_numeric(x, errors='coerce') and int(x) > 0 else 0)
df1 = df1.rename(columns={'baños': 'banos', 'estacionamientos': 'estacionamiento'})
df1
```

```python
df2 = pd.read_csv("/Users/dayth/OneDrive/Escritorio/Proyectos/EDA + ML/data/rawdata/pag2.csv", encoding="utf-8", delimiter=",")
df2
```

```python
cols = list(df2.columns)
precio_idx = cols.index('precio_uf')
df2.insert(precio_idx, 'precio', df2['precio_uf'] * 39240)
df2 = df2.drop(columns=['precio_uf', 'url', 'informacion_adicional', 'superficie_util'])
df2['estacionamiento'] = df2['estacionamiento'].replace({'Si': 1, 'No': 0})
df2
```

```python
df3 = pd.read_csv("/Users/dayth/OneDrive/Escritorio/Proyectos/EDA + ML/data/rawdata/pag3.csv", encoding="utf-8", delimiter=";")
df3 = df3.rename(columns={
    'Precio': 'precio',
    'Superficie Construida': 'superficie_construida',
    'Superficie Total': 'superficie_total',
    'Estacionamiento': 'estacionamiento',
    'Latitud': 'latitud',
    'Longitud': 'longitud',
    'Banos': 'banos',
    'Piezas': 'dormitorios',
    'Lugar': 'lugar'
})

df3['estacionamiento'] = df3['estacionamiento'].fillna(0)
df3['estacionamiento'] = df3['estacionamiento'].apply(lambda x: 1 if pd.to_numeric(x, errors='coerce') and int(x) > 0 else 0)

df3
```

```python
def fix_coord(val):
    try:
        s = str(val).replace('.', '').strip()
        if not s or s.lower() in ['nan', 'none']:
            return None

        # Si comienza con "-", conserva el signo y usa los dos dígitos siguientes
        if s.startswith('-'):
            num = s[1:]  # Elimina el signo para insertar el punto
            return float('-' + num[:2] + '.' + num[2:])
        else:
            return float(s[:2] + '.' + s[2:])
    except:
        return None

df3['latitud'] = df3['latitud'].apply(fix_coord)
df3['longitud'] = df3['longitud'].apply(fix_coord)
```

```python
df3 = df3[df3['Corredora'].isnull()]
df3 = df3.drop(columns=["lugar", "Titulo", "Fecha", 'Corredora'])
df3
```

```python
common_cols = set(df1.columns) & set(df2.columns) & set(df3.columns)
common_cols = list(common_cols)

df1_aligned = df1[common_cols]
df2_aligned = df2[common_cols]
df3_aligned = df3[common_cols]

df_merged = pd.concat([df1_aligned, df2_aligned, df3_aligned], ignore_index=True)

desired_order = [
    'precio',
    'banos',
    'dormitorios',
    'superficie_total',
    'superficie_construida',
    'estacionamiento',
    'latitud',
    'longitud'
]

df = df_merged[desired_order]

df
```
